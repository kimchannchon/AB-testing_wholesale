```python
import pandas as pd

# Load the dataset
file_path = 'Wholesale customers data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Step 1: Data Preparation - Calculate total sales for each customer
data['Total_Sales'] = data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].sum(axis=1)

# Display the first few rows to verify the new column
data.head()
```

```python
from scipy import stats

# Step 2: Grouping the data
group_A = data[data['Channel'] == 1]['Total_Sales']  # New marketing strategy
group_B = data[data['Channel'] == 2]['Total_Sales']  # Usual business practices

# Step 3: Metric Calculation - Calculate the average total sales for each group
avg_sales_A = group_A.mean()
avg_sales_B = group_B.mean()

# Step 4: Statistical Testing - Perform a t-test
t_stat, p_value = stats.ttest_ind(group_A, group_B)

avg_sales_A, avg_sales_B, t_stat, p_value
```
